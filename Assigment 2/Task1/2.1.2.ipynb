```
'''
Implement List comprehensions to produce the following lists.
Write List comprehensions to produce the following Lists
2.1 ['A', 'C', 'A', 'D', 'G', 'I', ’L’, ‘ D’]
2.2 ['x', 'xx', 'xxx', 'xxxx', 'y', 'yy', 'yyy', 'yyyy', 'z', 'zz', 'zzz', 'zzzz']
2.3 ['x', 'y', 'z', 'xx', 'yy', 'zz', 'xxx', 'yyy', 'zzz', 'xxxx', 'yyyy', 'zzzz']
2.4 [[2], [3], [4], [3], [4], [5], [4], [5], [6]]
2.5 [[2, 3, 4, 5], [3, 4, 5, 6], [4, 5, 6, 7], [5, 6, 7, 8]]
2.6 [(1, 1), (2, 1), (3, 1), (1, 2), (2, 2), (3, 2), (1, 3), (2, 3), (3, 3)]
'''
```

```
#2.1
word = "ACADGILD"
alphabet_list = [ alphabet for alphabet in word ]
```

```
#2.2
input_list = ['x','y','z']
result1 = [ item*num for item in input_list for num in range(1,5)  ]
print(str(result1))
```

```
#2.3
# Compress above for loop into a single list comprehension using technique [i <Upper for condition> <lower for condition>]
input_list = ['x','y','z']
result2 = [ item*num for num in range(1,5) for item in input_list  ]
print("['x','y','z'] => " +   str(result2))
```

```
#2.4
input_list = [2,3,4]
result3 = [ [item+num] for item in input_list for num in range(0,3)]
print(str(result3))
```

```
#2.5
input_list = [2,3,4,5]
result4 = [ [item+num for item in input_list] for num in range(0,4)  ]
print(str(result4))
```

```
#2.6
input_list=[1,2,3]
result5 = [ (b,a) for a in input_list for b in input_list]
print(str(result5))
```
