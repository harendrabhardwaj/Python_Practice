```
'''Write a Python function which takes a character (i.e. a string of length 1) and returns True if it is a vowel, False otherwise.'''
```

```
def isvowel(strg):
  if strg=="a" or strg=="e" or strg=="i" or strg=="o" or strg=="u":
    return True
  else:
    return False
```

```
strg = input()
if len(strg)>1:
  print("Invalid String")
else:
  ans = isvowel(strg)
  print(ans)
```
